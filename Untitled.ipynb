# Parsing the variants.csv file for all C321 clones

```python
data = pd.read_csv("variants.csv",dtype = {'IS_HET': str, 'AO': str, 'QA': str} )
```

```python
data.sort_values(by='EXPERIMENT_SAMPLE_LABEL')
data
```

```python
data = data[data.EXPERIMENT_SAMPLE_LABEL != 'W62']
data = data[data.EXPERIMENT_SAMPLE_LABEL != 'W72']
del data['CHROMOSOME']
data
```

Figure out a bit of stuff about the variant set....looking for numbers of certain types of variants, read depth, etc.

Instance 1: Number of variant calls with read depth < min_read_depth
Instance 2: Number of variant calls with allele fracgion < min_allele_fraction 

```python
min_read_depth = 10
min_allele_fraction = 0.7
variants = data[data.ALT.notnull()]
print("total variant calls: ",variants.UID.count())
print("variant calls under min read depth: ",variants[variants.DP < min_read_depth].UID.count())
print("variant calls under min allele fraction: ",variants[variants.AF < min_allele_fraction].UID.count())
```

Begin to cull the data.

Remove everything without a minimal read depth or allele fraction

```python
vc = variants[variants.DP > min_read_depth]
vc = vc[vc.AF > min_allele_fraction]
print("total culled variant calls: ",vc.UID.count())
```

Separate the culled variant calls into ECNR2.1, C321.A, and C321.A_F11 bins

```python
Evc = vc[vc.EXPERIMENT_SAMPLE_LABEL.str[0] == 'E']
Cvc = vc[vc.EXPERIMENT_SAMPLE_LABEL.str[0] == 'C']
Fvc = vc[vc.EXPERIMENT_SAMPLE_LABEL.str[0] == 'F']
print("total ENCR2.1 culled variant calls: ",Evc.UID.count())
print("total C321.A culled variant calls: ",Cvc.UID.count())
print("total C321.A_F11 culled variant calls: ",Fvc.UID.count())
```
